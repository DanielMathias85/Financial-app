```python
df['Estimated Shares Outstanding'].fillna(df['Estimated Shares Outstanding'].median(),inplace=True)
```

```python
plt.hist(df['For Year'])
plt.show()
```

```python
import matplotlib.pyplot as plt

# Remove NaN values before creating the histogram
plt.hist(df['Earnings Per Share'].dropna())
plt.show()
```

```python
df['Earnings Per Share'].fillna(df['Earnings Per Share'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn import metrics
```

```python
y = df['Estimated Shares Outstanding']
```

```python
y.head()
```

```python
X = df.drop(['Estimated Shares Outstanding'],axis=1)
X.head()
```

```python
# split the dataset into train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
test_data_prediction = model.predict(X_test)
score = metrics.r2_score(y_test,test_data_prediction)
print(score)
```
